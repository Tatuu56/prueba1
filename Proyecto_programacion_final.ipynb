2)Había columnas a las que le faltaban datos. Las columnas “InicioVisitaReal” y “FinVisitaReal” les faltaban algunos datos de pedidos que todavía no se habían puesto en marcha.
```python
# Crear un diccionario con la dirección como clave y la longitud/latitud como valor
direccion_coord = {}
for index, row in df.iterrows():
    if not pd.isnull(row['latitud']) and not pd.isnull(row['longitud']):
        direccion_coord[row['direccion']] = (row['latitud'], row['longitud'])

# Rellenar los valores faltantes
for index, row in df.iterrows():
    if pd.isnull(row['latitud']) and pd.isnull(row['longitud']):
        if row['direccion'] in direccion_coord:
            df.loc[index, ['latitud', 'longitud']] = direccion_coord[row['direccion']]
        else:
            df.loc[index, ['latitud']] = df['latitud'].mean()
            df.loc[index, ['longitud']] = df['longitud'].mean()
```

**Teníamos 42 pedidos que no incluían información de latitud y longitud.
Para resolver los outliers, seguimos este proceso:**

a)Revisamos cada pedido con datos faltantes para ver si su dirección coincidía con la de otros pedidos que sí tenían la latitud y longitud.

b)Si encontrábamos un pedido con la misma dirección que tenía los datos completos, tomábamos los valores de latitud y longitud de ese pedido y los asignábamos al pedido que carecía de ellos.

c)Si no encontrábamos ningún otro pedido con la misma dirección, rellenábamos los datos faltantes utilizando la media de latitud y longitud de todos los pedidos que sí tenían esta información.

Luego, buscamos identificar que porcentaje de clientes que hacen pedidos a la empresa regularmente. Para esto, llamamos **clientes constantes** a aquellos clientes que hicieron más de 5 pedidos a alguno de los clientes de la empresa en los últimos 3 meses. Esto es muy útil para la empresa para identificar la cantidad de pedidos que van a tener como base, y asi podrán ir midiendo su progreso. Mientras más clientes constantes, mayor regularidad en los pedidos y mayor número de personas que confía en su servicio.

```python
def clientes_constantes(df):
  total=0
  pedidos_por_cliente=df['iddomicilioorden'].value_counts()
  for x in pedidos_por_cliente:
    if x>5:
       total=total+1
  return total
clientes_constantes(df)
```

```python
!pip install haversine
from haversine import haversine

def calcular_distancia_total(df, cliente):
    # Filtra el DataFrame por cliente
    df_cliente = df[df['cliente'] == cliente].copy()

    # Ordena el DataFrame por la columna "InicioVisitaReal"
    df_cliente.sort_values(by=['InicioVisitaReal'], inplace=True)

    # Calcula la distancia entre cada par de puntos consecutivos
    distancias = []
    for i in range(len(df_cliente) - 1):
        punto1 = (df_cliente.iloc[i]['latitud'], df_cliente.iloc[i]['longitud'])
        punto2 = (df_cliente.iloc[i + 1]['latitud'], df_cliente.iloc[i + 1]['longitud'])
        distancia = haversine(punto1, punto2)  # Distancia en kilómetros
        distancias.append(distancia)

    # Suma las distancias para obtener la distancia total
    distancia_total = sum(distancias)

    return distancia_total

distancia_cliente_20 = calcular_distancia_total(df, 20)
distancia_cliente_70 = calcular_distancia_total(df, 70)

print(f"Distancia total para el cliente 20: {distancia_cliente_20:.2f} km")
print(f"Distancia total para el cliente 70: {distancia_cliente_70:.2f} km")
```

```python
!pip install haversine
from haversine import haversine

def calcular_distancia_total_por_dia(df, cliente):
    # Filtra el DataFrame por cliente
    df_cliente = df[df['cliente'] == cliente].copy()

    # Agrupa por día
    distancias_por_dia = {}
    for dia, grupo in df_cliente.groupby(df_cliente['InicioVisitaReal'].dt.date):
        # Ordena el grupo por la columna "InicioVisitaReal"
        grupo.sort_values(by=['InicioVisitaReal'], inplace=True)

        # Calcula la distancia entre cada par de puntos consecutivos
        distancias = []
        for i in range(len(grupo) - 1):
            punto1 = (grupo.iloc[i]['latitud'], grupo.iloc[i]['longitud'])
            punto2 = (grupo.iloc[i + 1]['latitud'], grupo.iloc[i + 1]['longitud'])
            distancia = haversine(punto1, punto2)  # Distancia en kilómetros
            distancias.append(distancia)

        # Suma las distancias para obtener la distancia total del día
        distancia_total_dia = sum(distancias)

        # Agrega la distancia total del día al diccionario
        distancias_por_dia[dia] = distancia_total_dia

    return distancias_por_dia

distancias_cliente_20 = calcular_distancia_total_por_dia(df, 20)
distancias_cliente_70 = calcular_distancia_total_por_dia(df, 70)

print("Distancias para el cliente 20 por día:")
for dia, distancia in distancias_cliente_20.items():
    print(f"{dia}: {distancia:.2f} km")

print("\nDistancias para el cliente 70 por día:")
for dia, distancia in distancias_cliente_70.items():
    print(f"{dia}: {distancia:.2f} km")
```

## **Conclusiones finales**

**Tiempos de entrega**

Como ya vimos, los tiempos de entrega varían dependiendo del cliente. El cliente 20 tiene tiempos más largos de entrega, y esto se da porque son generalmente pedidos más pesados, y porque las zonas de entrega están más dispersas entre sí. Este segundo argumento hace que se pierda tiempo yendo a entregar pedidos muy alejados de las zonas en la que Iflow suele repartir. Los tiempos de entrega del cliente 70 son más bajos porque los pedidos son más livianos y los puntos de entrega están muy cerca entre sí, por lo que optimiza mucho tiempo.

**Solución**

Para manejar los pedidos más pesados, Iflow podría asignar vehículos de mayor capacidad o aumentar la cantidad de entregas en cada viaje. Incluso podrían establecerse "ventanas de entrega" para este tipo de pedidos, permitiendo mayor flexibilidad en el cronograma y la posibilidad de realizar entregas en bloque en momentos estratégico. Repartir en la noche o madrugada estos pedidos alejados reduciría los tiempos de entrega y los costos, dando una mayor satisfacción al cliente. Por consecuencia, esto podría aumentar la cantidad de clientes constantes de la empresa.

**Distancias lejanas de clientes puntuales**

Otro problema que notamos al analizar todos los casos fue la lejanía de algunos pedidos que son muy puntuales, y representan menos del 1% de los pedidos. Aunque sean pocos pedidos, esto puede repercutir en la empresa ya que puede hacer que se pierda plata realizando esas entregas largas, y se podría perder rentabilidad por un detalle que se puede corregir.

**Solución**

La solución a este problema puede ser establecer un mínimo de pedidos para realizar entregas en zonas alejadas, incentivando al cliente a pedir un mínimo de pedidos o unidades para que se pueda realizar el envío. Otra alternativa sería poner una tarifa adicional basada en la distancia para estos casos específicos para cubrir los costos extras, asegurando que estas entregas no afecten la rentabilidad.
