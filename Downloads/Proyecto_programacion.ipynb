```python
# Crear un diccionario con la direccin como clave y la longitud/latitud como valor
direccion_coord = {}
for index, row in df.iterrows():
    if not pd.isnull(row['latitud']) and not pd.isnull(row['longitud']):
        direccion_coord[row['direccion']] = (row['latitud'], row['longitud'])

# Rellenar los valores faltantes
for index, row in df.iterrows():
    if pd.isnull(row['latitud']) and pd.isnull(row['longitud']):
        if row['direccion'] in direccion_coord:
            df.loc[index, ['latitud', 'longitud']] = direccion_coord[row['direccion']]
        else:
            df.loc[index, ['latitud']] = df['latitud'].mean()
            df.loc[index, ['longitud']] = df['longitud'].mean()
```

```python

orders_by_client_date = df.groupby(['cliente', df['InicioVisitaReal'].dt.date])['iddomicilioorden'].count().reset_index()


client_70_data = orders_by_client_date[orders_by_client_date['cliente'] == 70]
client_20_data = orders_by_client_date[orders_by_client_date['cliente'] == 20]


plt.figure(figsize=(10, 6))

plt.plot(client_20_data['InicioVisitaReal'], client_20_data['iddomicilioorden'], label='Cliente 20')
plt.plot(client_70_data['InicioVisitaReal'], client_70_data['iddomicilioorden'], label='Cliente 70')

plt.xlabel('Fecha')
plt.ylabel('Cantidad de Pedidos')
plt.title('Pedidos por Da para Cliente 70 y Cliente 20')
plt.legend()
plt.grid(True)


plt.show()
```

La grafica muestra como el cliente 70 es mucho mas constante en el dia a dia que el cliente 20 y podemos concluir que esto se debe a su mayor base de clientes constantes (que hacen mas de 5 pedidos en 3 meses). Esto podra permitirle a la empresa planificar ciertos repartidores para el cliente 70 que ya se sabe que va a haber mas o menos la misma cantidad de pedidos cada da, y tener otros repartidores que sean mas variables dependiendo el dia, para cubrir la gran variacin en la demanda del cliente 20.
